# **1. Import des bibliothèques nécessaires**

```python
import pandas as pd  # Pour la manipulation des données\import numpy as np  # Pour les opérations mathématiques
import matplotlib.pyplot as plt  # Pour la création de graphiques
import seaborn as sns  # Pour des visualisations avancées
from sklearn.model_selection import train_test_split  # Pour diviser les données en ensembles d'entraînement et de test
from sklearn.preprocessing import StandardScaler, OneHotEncoder  # Pour la mise à l'échelle et l'encodage des données
from sklearn.compose import ColumnTransformer  # Pour appliquer des transformations sur différentes colonnes
from sklearn.pipeline import Pipeline  # Pour chaîner les étapes de traitement et de modélisation
from sklearn.impute import SimpleImputer  # Pour gérer les valeurs manquantes
from sklearn.linear_model import LogisticRegression  # Algorithme de régression logistique
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve  # Pour évaluer les performances du modèle
import joblib # Pour sauvegarder et convertir le fichier en ".pkl"
```

# **2. Chargement des données**

```python
data_path = "../Consigne/bank-additional/bank-additional/bank-additional-full.csv"  # Chemin vers le fichier de données
data = pd.read_csv(data_path, sep=";")  # Charger les données CSV avec un séparateur ";"
```

# **3. Nettoyage des colonnes**

```python
# noms de colonnes : Supprimer les espaces  
data.columns = data.columns.str.strip() 
# Variable cible :
 # Voir si la var cible existe 
if 'y' not in data.columns: 
    raise ValueError("La colonne 'y' n'existe pas dans le DataFrame.")
 # La convertir en format binaire
data['y'] = data['y'].str.strip().map({'yes': 1, 'no': 0})  
```

# **4. Validation des colonnes**

```python
print("Colonnes disponibles dans le DataFrame :", data.columns) 
```

# **5. Visualisation exploratoire des données**

```python
# Distribution de la variable cible 'y'
sns.countplot(x=data['y'], palette='viridis')  
plt.title("Distribution de la variable cible (y)")
plt.xlabel("Acceptation de l'offre (0: Non, 1: Oui)")
plt.ylabel("Nombre de clients")
plt.show()
```

```python
# Visualisation des données numériques
num_columns = data.select_dtypes(include=['int64', 'float64']).columns  # Colonnes numériques
for col in num_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=30, color='blue')  
    plt.title(f"Distribution de {col}")
    plt.xlabel(col)
    plt.ylabel("Fréquence")
    plt.show()
```

```python
# Visualisation des données catégoriques
cat_columns = data.select_dtypes(include=['object']).columns  # Colonnes catégoriques
for col in cat_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=data[col], palette='Set2', order=data[col].value_counts().index)  
    plt.title(f"Distribution de {col}")
    plt.xlabel("Nombre de clients")
    plt.ylabel(col)
    plt.show()
```

```python
# Matrice de corrélation pour les colonnes numériques
plt.figure(figsize=(12, 10))
correlation_matrix = data[num_columns].corr()  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')  
plt.title("Matrice de corrélation des variables numériques")
plt.show()
```

```python
# Relation entre certaines variables et la cible 'y'
for col in ['age', 'duration', 'campaign']:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data['y'], y=data[col], palette='viridis')  
    plt.title(f"Relation entre {col} et la cible 'y'")
    plt.xlabel("Acceptation de l'offre (0: Non, 1: Oui)")
    plt.ylabel(col)
    plt.show()
```

# **6. Séparation des colonnes numériques et catégoriques**

```python
num_features = data.select_dtypes(include=['int64', 'float64']).columns.tolist()  # Liste des colonnes numériques
cat_features = data.select_dtypes(include=['object']).columns.tolist()  # Liste des colonnes catégoriques

print("Colonnes numériques spécifiées :", num_features)
print("Colonnes catégoriques spécifiées :", cat_features)
```

# **7. Division des données**

```python
X = data.drop('y', axis=1)  # Variables explicatives
y = data['y']  # Variable cible
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)  # Division en ensembles
```

# **8. Validation des colonnes dans X_train**

```python
num_features = [col for col in num_features if col in X_train.columns]  
cat_features = [col for col in cat_features if col in X_train.columns]  

print("Colonnes numériques après validation :", num_features)
print("Colonnes catégoriques après validation :", cat_features)
```

# **9. Création des transformations**

```python
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # Remplir les valeurs manquantes par la médiane
    ('scaler', StandardScaler())  
])
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Remplir les valeurs manquantes par la valeur la plus fréquente
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # Encodage des var cat
])
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),  
        ('cat', cat_transformer, cat_features)  
    ]
)
```

# **10. Validation des transformations**

```python
# Régression logistique comme classificateur : entrainemt modèle

try:
    X_train_transformed = preprocessor.fit_transform(X_train)  # Transformation des données d'entraînement
    print("Dimensions après transformation :", X_train_transformed.shape) 
except Exception as e:
    print("Erreur lors de la transformation :", e)  

# **Pipeline avec régression logistique**
model = Pipeline(steps=[
    ('preprocessor', preprocessor),  
    ('classifier', LogisticRegression())  
])
```

# **11. Entraînement du modèle**

```python
# Entraîner le modèle 

try:
    model.fit(X_train, y_train)  
    print("Modèle entraîné avec succès.")
except Exception as e:
    print("Erreur lors de l'entraînement :", e)  
```

# **12. Prédictions et évaluation**

```python
try:
    y_pred = model.predict(X_test)  
    y_pred_proba = model.predict_proba(X_test)[:, 1]  

    print("Rapport de classification:\n", classification_report(y_test, y_pred))  
    print("Matrice de confusion:\n", confusion_matrix(y_test, y_pred)) 

    # Courbe ROC
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)  
    roc_auc = roc_auc_score(y_test, y_pred_proba)  
    plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")  
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Courbe ROC")
    plt.legend()
    plt.show()
except Exception as e:
    print("Erreur lors de la prédiction :", e)  
```

# **13. Conversion du fichier pour le déploiement**

```python
# Sauvegarder le modèle entraîné
joblib.dump(model, "bank_marketing_model.pkl")  
print("Modèle sauvegardé avec succès.")
```
